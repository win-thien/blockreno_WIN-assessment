```python
#importing libraries and modules 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

#### Question 1: Which month has the highest count of valid users created?

```python
quest1 = pd.read_csv('Question 1.csv')
quest1
```

June has the highest count of valid users.

```python
plt.figure(figsize=(15,8))
plt.bar(quest1['month_created'], quest1['count'])
plt.title('Valid Users Signup')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```
